```python
import numpy as np
import matplotlib.pyplot as plt
import sinesum as ss
```

```python
#N Values
nVal = [1,3,5,10,30,100]
#A values
aVal = [0.01,0.25,0.49]
T = 2*np.pi
for i in range(4):
    for j in range(6):
        #Creates all the necessary plots
        f = plt.figure(figsize=(6,4))
        a = plt.axes()
        yVals = np.array([ss.Sn(aVal[i],a+1) for a in range(nVal[j]+1)])
        a.plot(np.arange(0,nVal[j]+1,1), yVals, color = "blue")
        a.set(xlabel= "iterations (n): "+str(nVal[j]), ylabel= ("a ="+str(aVal[i])), title="Sine sum")
        plt.show()
```
